```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#number of steps
T = 10000

#initial price
p0 = 10

#drift
mu = 0.0001
```

```python
#last log price
last_p = np.log(p0)

log_returns = np.zeros(T)
prices = np.zeros(T)
for t in range(T):
  #sample a log return
  r = 0.01 * np.random.randn()

  #compute the new log price
  p = last_p + mu + r

  #store return and price
  log_returns[t] = r + mu
  prices[t] = np.exp(p)

  last_p = p
```

```python
plt.figure(figsize = (20,8))
plt.plot(prices)
plt.title('Stimulated Price');
```
