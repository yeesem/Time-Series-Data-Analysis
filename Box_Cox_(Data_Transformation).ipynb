```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import boxcox
```

```python
df = pd.read_csv("/content/airline_passengers.csv",index_col = 'Month',parse_dates = True)
```

```python
df.head()
```

```python
#The number of passengers is increasing over the time
#The amplitude is increasing over the time
df['Passengers'].plot(figsize = (20,8));
```

```python
df['SqrtPassengers'] = np.sqrt(df['Passengers'])
```

```python
#The time series has been squashed down slightly but the amplitude of seasonal still increase over the time
df['SqrtPassengers'].plot(figsize = (20,8))
```

```python
df['LogPassengers'] = np.log(df['Passengers'])
```

```python
df['LogPassengers'].plot(figsize = (20,8))
```

```python
data,lam = boxcox(df['Passengers'])
```

```python
lam
```

```python
df['BoxCoxPassengers'] = data
df['BoxCoxPassengers'].plot(figsize = (20,8))
```
